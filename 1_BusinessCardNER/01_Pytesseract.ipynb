```python
import numpy as np
import pandas as pd
import cv2
import PIL
import pytesseract
import string
import re
```

```python
img_cv = cv2.imread('./Selected_Vie/002.jpeg')
img_pl = PIL.Image.open('./Selected_Vie/002.jpeg')
# cv2.imshow('Bussiness card', img_cv)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```

```python
data = pytesseract.image_to_data(img_gray, lang='vie')
```

```python
print(data)
```

```python
dataList = list(map(lambda x: x.split('\t'), data.split('\n')))
df = pd.DataFrame(dataList[1:], columns=dataList[0])
```

```python
df
```

```python
whitespace = string.whitespace
punctuation = "!#$%&\'()*+:;<=>?[\\]^`{|}~"
tableWhitespace = str.maketrans('','',whitespace)
tablePunctuation = str.maketrans('','',punctuation)
def cleanText(txt):
    text = str(txt)
    text = text.lower()
    removewhitespace = text.translate(tableWhitespace)
    removepunctuation = removewhitespace.translate(tablePunctuation)
    
    return str(removepunctuation)
```

```python
df['text'] = df['text'].apply(cleanText)
```

```python
dataClean = df.query("text != ''")
dataClean.dropna(inplace=True)
```

```python
df.dropna(inplace=True)
col_int = ['level', 'page_num', 'block_num', 'par_num', 'line_num', 'word_num', 'left', 'top', 'width', 'height', 'conf']
df[col_int] = df[col_int].astype(int)
```

```python
image_copy = img_cv.copy()
level = 'word'
for l, x, y, w, h, c, txt in df[['level', 'left', 'top', 'width', 'height', 'conf', 'text']].values:
    if level == 'page':
        if l == 1:
            cv2.rectangle(image_copy,(x,y),(x+w,y+h),(0,0,0),2)
        else:
            continue
    elif level == 'block':
        if l == 2:
            cv2.rectangle(image_copy,(x,y), (x+w,y+h),(255,0,0),2)
        else:
            continue
    elif level == 'para':
        if l == 3:
            cv2.rectangle(image_copy,(x,y), (x+w,y+h), (0,255,0),2)
        else:
            continue
    elif level == 'line':
        if l == 4:
            cv2.rectangle(image_copy,(x,y), (x+w,y+h), (0,0,255), 2)
        else:
            continue
    elif level == 'word':
        if l == 5:
            cv2.rectangle(image_copy, (x,y), (x+w,y+h), (0,255,0),2)
            # cv2.putText(image_copy,txt,(x,y),cv2.FONT_HERSHEY_PLAIN,1,(255,0,0),2)
        else:
            continue
cv2.imshow('Bounding box', image_copy)
cv2.waitKey()
cv2.destroyAllWindows()
```
