```python
import numpy as np
import pandas as pd
import string
import re
```

```python
with open('business_card_data_vie_clean.txt', mode='r',encoding='utf8',errors='ignore') as f:
    text = f.read()
```

```python
data = list(map(lambda x:x.split('\t'), text.split('\n')))
```

```python
df = pd.DataFrame(data[1:],columns=data[0])
```

```python
df.head(20)
```

```python
whitespace = string.whitespace
punctuation = "!#$%&\'()*+:;<=>?[\\]^`{|}~"
tableWhitespace = str.maketrans('','',whitespace)
tablePunctuation = str.maketrans('','',punctuation)
def cleanText(txt):
    text = str(txt)
    text = text.lower()
    removewhitespace = text.translate(tableWhitespace)
    removepunctuation = removewhitespace.translate(tablePunctuation)
    
    return str(removepunctuation) 
```

```python
df['text'] = df['text'].apply(cleanText)
```

```python
dataClean = df.query("text != ''")
dataClean.dropna(inplace=True)
```

```python
group = dataClean.groupby(by='id')
```

```python
cards = group.groups.keys()
```

```python
allCardsData = []
for card in cards:
    cardData = []
    grouparray = group.get_group(card)[['text','tag']].values
    content = ''
    annotations = {'entities':[]}
    start = 0
    end = 0
    for text, label in grouparray:
        text = str(text)
        stringLength = len(text) + 1

        start = end
        end = start + stringLength

        if label != 'O':
            annot = (start,end-1,label)
            annotations['entities'].append(annot)

        content = content + text + ' '
        
        
    cardData = (content,annotations)
    allCardsData.append(cardData)
```

```python
allCardsData
```

```python
import random
```

```python
random.shuffle(allCardsData)
```

```python
split_index = int(len(allCardsData) * 0.8)
TrainData = allCardsData[:split_index]
TestData = allCardsData[split_index:]
```
